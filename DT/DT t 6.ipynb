```python
A = Matrix([
    [2, 3, 1],
    [2, 0, -1],
    [-2, 1, -2]
])
```

```python
A
```

```python
A.det()
```

$$x'(t) = A \cdot x(t)\\x(0)=\left[0,1,-2\right]$$

```python
lambdas = [ v[0] for v in A.eigenvects() ]
```

```python
lambdas = [ simplify(l) for l in lambdas ]
lambdas
```

```python
t = symbols('t')
Lambda = Matrix([
    [ exp(lambdas[0]*t), 0, 0],
    [ 0, exp(lambdas[1]*t), 0],
    [ 0, 0, exp(lambdas[2]*t)]
])
Lambda
```

```python
#V = Matrix([ v[2][0].T for v in A.eigenvects() ]).T
AE=A.eigenvects()
V = Matrix.hstack( AE[0][2][0],AE[1][2][0],AE[2][2][0]) 
V
```

```python
for i in range(3):
    for j in range(3):
        V[i,j] = simplify(V[i,j])
V
```

```python
Vi=V.inv()
Vi
```

```python
x0 = Matrix([
    [0],
    [1],
    [-2]
])
```

```python
x0
```

```python
X = V * Lambda * Vi * x0
X
```
