### Problem 1
<p style="text-align: right">(12 points)</p>

Write a program that asks two strings from the user and then prints the number of bigrams that appear in both strings.
By bigram we mean a substring of length 2, for instance the bigram decomposition of the string `papabear` is `pa`-`ap`-`pa`-`ab`-`be`-`ea`-`ar`. Example run:

```
1st string: papabear
2nd string: mamabear
Number of common bigrams: 4
```

````python
in1 = input('Enter first string:  ')
in2 = input('Enter secon string:  ')
def bigram(inn):
    s = set()
    for i in range(len(inn)-1):
        s.add((inn[i]+inn[i+1]))
    return s

s1 = bigram(in1)
s2 = bigram(in2)
cnt = 0

#for c in s1:
#    if c in s2:
#        cnt += 1

cnt = len(s1&s2)

print (f'Number of common bigrams: {cnt}')
````

````python
# solution 2
def bigrams(s):
    return set(zip(s[:-1], s[1:]))

s1 = input('1st string: ')
s2 = input('2nd string: ')
print('Number of common bigrams:', len(bigrams(s1) & bigrams(s2)))
````
