```python
# linear regression
N = 400
x = np.random.normal(0,2,N)
y = 3*x+5
noise = np.random.normal(1,10,N)
y = y+noise
```

```python
# coefficients for regression line
a = (np.mean(x*y)-np.mean(x)*np.mean(y))/np.var(x)
b = np.mean(y) - a*np.mean(x)
```

```python
# correlation
r = np.corrcoef(x,y)[1,0]
r
```

```python
# plot the points
plt.plot(x,y,'.')
x_r = np.arange(-10,10,0.01)
plt.plot(x_r, a * x_r + b,'r', label=f'regression line, r = {round(r,4)}')
plt.legend()
plt.show()
```
