```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = stats.binom(10,0.2)
m = np.mean(X.rvs(100))
m
```

```python
X = stats.binom(100,0.7)
k = np.arange(0,101,1)
y = X.pmf(k)
plt.bar(k, y)
plt.show()
```

```python
X = stats.binom(10,0.5)
k = np.arange(0,11,1)
y = X.pmf(k)
plt.bar(k, y)
plt.show()
```

```python
Y = stats.poisson(2)
print (Y.pmf(3))
print (Y.rvs())
```

```python
Y = stats.poisson(3)
k = np.arange(0,20,1)
y = Y.pmf(k)
plt.bar(k, y)
plt.show()
```

```python
Y = stats.poisson(20)
k = np.arange(0,20,1)
y = Y.pmf(k)
plt.bar(k, y)
plt.show()
```

```python
Y = stats.geom(0.75)
print (Y.pmf(3))
print (Y.rvs())
```

```python
Y = stats.geom(0.2)
k = np.arange(0,20,1)
y = Y.pmf(k)
plt.bar(k, y)
plt.show()
```

```python
Y1 = stats.norm(3,2)
Y2 = stats.norm(3,4)
Y3 = stats.norm(3,6)

k = np.arange(-12,16,0.01)

y1 = Y1.pdf(k)
y2 = Y2.pdf(k)
y3 = Y3.pdf(k)

plt.plot(k, y1, label='$\mu=3$, $\sigma=2$')
plt.plot(k, y2, label='$\mu=3$, $\sigma=4$')
plt.plot(k, y3, label='$\mu=3$, $\sigma=6$')
plt.title('PDFs of various normal distribution ')
plt.legend()
plt.show()
```

```python
Y1 = stats.norm(3,2)
Y2 = stats.norm(3,4)
Y3 = stats.norm(3,6)

k = np.arange(-12,16,0.01)

y1 = Y1.cdf(k)
y2 = Y2.cdf(k)
y3 = Y3.cdf(k)

plt.plot(k, y1, label='$\mu=3$, $\sigma=2$')
plt.plot(k, y2, label='$\mu=3$, $\sigma=4$')
plt.plot(k, y3, label='$\mu=3$, $\sigma=6$')
plt.title('CDFs of various normal distribution ')
plt.legend()
plt.show()
```
