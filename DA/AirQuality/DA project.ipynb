```python
# Sand
import csv
import tkinter as tk
from tkinter import filedialog, messagebox
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def read_data(filename):
    data = []
    with open(filename, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=';')
        header = next(reader)
        for row in reader:
            record = {
                'Date': row[0],
                'Time': row[1],
                'CO(GT)': float(row[2].replace(',', '.')) if row[2] else None,
                'PT08.S1(CO)': int(row[3]) if row[3] else None,
                'NMHC(GT)': int(row[4]) if row[4] else None,
                'C6H6(GT)': float(row[5].replace(',', '.')) if row[5] else None,
                'PT08.S2(NMHC)': int(row[6]) if row[6] else None,
                'NOx(GT)': int(row[7]) if row[7] else None,
                'PT08.S3(NOx)': int(row[8]) if row[8] else None,
                'NO2(GT)': int(row[9]) if row[9] else None,
                'PT08.S4(NO2)': int(row[10]) if row[10] else None,
                'PT08.S5(O3)': int(row[11]) if row[11] else None,
                'T': float(row[12].replace(',', '.')) if row[12] else None,
                'RH': float(row[13].replace(',', '.')) if row[13] else None,
                'AH': float(row[14].replace(',', '.')) if row[14] else None,
            }
            data.append(record)
    return data

def analyze_data(filename):
    data = read_data(filename)
    df = pd.DataFrame(data)
    df = df.fillna(0)

    selected_columns = ['CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 'PT08.S2(NMHC)',
                        'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']

    df[selected_columns] = df[selected_columns].apply(pd.to_numeric, errors='coerce')

    plt.figure(figsize=(12, 8))
    for column in selected_columns[:11]:
        plt.plot(df['Date'] + ' ' + df['Time'], df[column], label=column)

    plt.title("Variations of Chemicals Over Time")
    plt.xlabel("Date and Time")
    plt.ylabel("Chemical Concentration")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()


    describe_stats = df[selected_columns].describe()
    mean_values = df[selected_columns].mean()
    median_values = df[selected_columns].median()
    variance_values = df[selected_columns].var()
    correlation_matrix = df[selected_columns].corr()


    threshold = 5.0
    pollutant_column = 'CO(GT)'
    exceeding_threshold = df[df[pollutant_column] > threshold]
    frequency = len(exceeding_threshold) / len(df) * 100

    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(df[selected_columns])
    pca_df = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])

    return describe_stats, mean_values, median_values, variance_values, correlation_matrix, frequency, pca_df

def browse_file():
    file_path = filedialog.askopenfilename(filetypes=[("CSV Files", "*.csv")])
    if file_path:
        describe_stats, mean_values, median_values, variance_values, correlation_matrix, frequency, pca_df = analyze_data(file_path)

        frequency_text = f"Percentage of CO(GT) values exceeding 5.0: {frequency:.2f}%"
        messagebox.showinfo("Frequency Analysis Result", frequency_text)

        plt.figure(figsize=(10, 6))
        plt.scatter(pca_df['PCA1'], pca_df['PCA2'])
        plt.title("PCA Result")
        plt.xlabel("PCA1")
        plt.ylabel("PCA2")
        plt.show()


root = tk.Tk()
root.title("Air Quality Data Analysis")

browse_button = tk.Button(root, text="Browse CSV File", command=browse_file)
browse_button.pack()

root.mainloop()
```
