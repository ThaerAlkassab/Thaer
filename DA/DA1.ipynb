```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(np.random.randint(0,2,10))
plt.show()
```

```python
N=1000
values=np.random.randint(0,2,size=N)
trails=np.arange(1,N+1)
rel_freq=np.cumsum(values)/trails 
plt.plot(trails, rel_freq)
plt.ylim(0,1)
plt.show()
```

```python
N=1000
values=np.random.choice([1,2], p=[0.6,0.4],size=N)
trails=np.arange(1,N+1)
values= (values==1).astype(int)
rel_freq=np.cumsum(values)/trails 
plt.plot(trails, rel_freq)
plt.ylim(0,1)
plt.show()
```
