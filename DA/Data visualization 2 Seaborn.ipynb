```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
anscombe = sns.load_dataset('anscombe')
```

### Example:
Plot the Anscombe datasets and show the regression line using [sns.lmplot](https://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot)

```python
sns.lmplot(
    data=anscombe, x="x", y="y", col="dataset", hue="dataset",
    col_wrap=4, palette="muted", ci=None,
    height=4, scatter_kws={"s": 100, "alpha": 1}
)
```

```python
sns.lmplot(
    data=anscombe, x="x", y="y", col="dataset", hue="dataset",
    col_wrap=4, palette="muted", ci=None,
    height=4, scatter_kws={"s": 100, "alpha": 1}, robust = True
)
```
