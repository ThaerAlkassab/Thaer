**Exercice 1.**

Consider 2D heat transfer across the domain shown in the figure below with  H = W = 20.0 and D = 6.0. Solve the heat transfer problem, with the following boundary conditions:

• The external part of the boundary is convecting to 300K with a heat transfer coefficient of 750 W/(m$^2$K).

In the domain use the following material property:

• The thermal conductivity:

$$
k(x,y)=\begin{cases}
k_0 = 96 W/m·K, & \quad \text{ in } \Omega_2 ,\\
k_1 = 296 W/m·K, & \quad \text{ in }  \Omega_1),
\end{cases}
$$

and the load:

$$ f(x, y) = \begin{cases}
0.0, & \quad \text{ in }  \Omega_2,\\
100, & \Omega_1.
\end{cases}$$

image.png

```python
from dolfin import *
import matplotlib.pyplot as plt
from mshr import *
#
# to plot the result in the Notebook
%matplotlib inline
#
d = 1
#
H, W = 20.0, 20.0
domain = Rectangle(Point( 0, 0 ),Point ( W, H ) )
circ_rad = 3.0
circ = Circle(Point(10.0, 10.0),circ_rad)
domain.set_subdomain(1, circ)
mesh = generate_mesh (domain ,50)

Vh  = FunctionSpace(mesh, 'P', 2)

h  = Constant(750.0)
T0 = Constant(300.0) # for external shape
kappa0 = Constant(96.)
kappa1 = Constant(296.)
f0 = Expression('0.0', degree=d+2)
f1 = Expression('100.0', degree=d+2)
#
# Define boundary labels
#
class LeftBoundary(SubDomain):
  def inside(self, x, on_boundary):
    return on_boundary and near(x[0] , 0)

class RightBoundary(SubDomain):
  def inside(self, x, on_boundary):
    return on_boundary and near(x[0] - 20.0, 0)

class BottomBoundary(SubDomain):
  def inside(self, x, on_boundary):
    return on_boundary and near(x[1] , 0)

class TopBoundary(SubDomain):
  def inside(self, x, on_boundary):
    return on_boundary and near(x[1] - 20.0, 0)
#
class Inner_Cercle(SubDomain):
  def inside(self , x , on_boundary):
      r = sqrt((x[0]-10)**2 + (x[1]-10)**2)
      return (on_boundary) and (r < 3.0 * 1.0001)

leftBoundary = LeftBoundary()
rightBoundary = RightBoundary()
topBoundary = TopBoundary()
bottomBoundary = BottomBoundary()
inner_Cercle = Inner_Cercle()

# Initialize mesh function for boundary domains
boundaries = MeshFunction("size_t", mesh, mesh.topology().dim()-1, 0)
#
# mark the entire boundary as 0
#
boundaries.set_all(0)
#
# Mark the isolated parts of the boundaries as subdomains  1
#
rightBoundary.mark(boundaries, 1)
leftBoundary.mark(boundaries, 1)
topBoundary.mark(boundaries, 1)
bottomBoundary.mark(boundaries, 1)
#
inner_Cercle.mark(boundaries, 2)
#
#  ds: surface integration
#
ds=Measure('ds', domain=mesh, subdomain_data=boundaries)
#
#  weak form
u = TrialFunction(Vh)
v = TestFunction(Vh)
#
a = kappa0*dot(grad(u), grad(v))*dx + kappa1*dot(grad(u), grad(v))*dx + h*u*v*ds(1)
L =  f1*v*dx + h*T0*v*ds(1)
#
uh = Function(Vh)
#
# Solution
solve(a==L, uh)
#
plt.figure(figsize=(8,8))
p = plot(uh)
plt.title('Solution (K)')
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.colorbar(p)
```
