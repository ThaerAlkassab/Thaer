#### [copy](https://docs.python.org/3/library/copy.html)
- Contains a shallow and a deep copy function.

```python
import copy
```

```python
# In Python, assignment does NOT copy, it only creates a reference.
a = [10, 20, [30, 40]]
b = a
b[0] = 999
print (a)
print (b)
```

```python
# Making a shallow copy.
# In Python, assignment does NOT copy, it only creates a reference.
a = [10, 20, [30, 40]]
b = copy.copy(a)
b[0] = 999
print (a)
print (b)
```

```python
# ...but:
a = [10, 20, [30, 40]]
b = copy.copy(a)
b[2][0] = 999
print (a)
print (b)
```

Shallow copy only copies at the highest level of the data structure!

```python
# Making a deep copy of a list of lists object.
a = [10, 20, [30, 40]]
b = copy.deepcopy(a)
b[2][0] = 999
print (a)
print (b)
```

```python
# Two handy utility functions.

def to_pickle(obj, fname, protocol=4):
    '''Serialize object to file.'''
    pickle.dump(obj, open(fname, 'wb'), protocol)
    
def from_pickle(fname):
    '''Deserialize object from file.'''
    return pickle.load(open(fname, 'rb'))
```

```python
to_pickle(data, 'data.pkl')
```

```python
from_pickle('data.pkl')
```

## Exercise: Conway's Game of Life

Write a program that implements [Conway's Game of Life](https://en.wikipedia.org/wiki/Conway%27s_Game_of_Life)!

- <2 or >3 live neighbors: cell dies
- 3 live neigbors: cell will be living
- every cell has 8 neigbors
- the status of the border cells never changes

```python
# Let's specify the initial state in a string!
# (We could read it from file too, but let's just use a string for simplicity!)

worldstr = '''
....................
....................
....................
....................
....................
....................
....................
....................
.........oo.........
........oo..........
.........o..........
....................
....................
....................
....................
....................
....................
....................
....................
....................
'''.strip()
```

```python
import numpy as np
import copy
# Load intial state to NumPy array

world = np.array([list(l) for l in worldstr.split('\n')])
```

```python
def dis_world(world):
    for x in world:
        print(' '.join(x))

dis_world(world)
```

```python
i = 10
j = 9
world[i - 1:i + 2, j - 1:j + 2]
```

```python
def count_naighbors(world):
    nrows, ncols = world.shape
    nnbs = np.zeros(world.shape, dtype='uint8')
    for i in range(1, nrows - 1):
        for j in range(1, ncols - 1):
            cnt = (world[i - 1:i + 2, j - 1:j + 2] == 'o').sum() - (world[i, j] == 'o')
            nnbs[i, j] = cnt
    return nnbs
```

```python
def update_world(nbs):
    nrows, ncols = nbs.shape
    nnbs = copy.copy (world)
    for i in range(1, nrows - 1):
        for j in range(1, ncols - 1):
            if (nbs[i,j]>3) or (nbs[i,j]<2):
                nnbs[i,j] = '.'
            elif nbs[i,j] == 3:
                nnbs[i,j] = 'o'
    return nnbs
```

```python
while True:
    dis_world(world)
    input()
    nbs = count_naighbors(world)
    world = update_world(nbs)
```

```python
worldstr = '''
....................
....................
....................
....................
....................
....................
....................
....................
.........oo.........
........oo..........
.........o..........
....................
....................
....................
....................
....................
....................
....................
....................
....................
'''.strip()

import time
import numpy as np
from IPython.display import clear_output

def display_world(world):
    for x in world:
        print(' '.join(x))

def count_neighbors(world):
    nrows, ncols = world.shape
    nnbs = np.zeros(world.shape, dtype='uint8')
    for i in range(1, nrows - 1):
        for j in range(1, ncols - 1):
            cnt = (world[i - 1:i + 2, j - 1:j + 2] == 'o').sum() - (world[i, j] == 'o')
            nnbs[i, j] = cnt
    return nnbs
        
# load initial state to NumPy array.
world = np.array([list(x) for x in worldstr.split('\n')])

# main loop
while True:
    clear_output(wait=True)
    display_world(world)
    time.sleep(0.5)
    nnbs = count_neighbors(world)
    world[nnbs == 3] = 'o'               # reproduction
    world[(nnbs < 2) | (nnbs > 3)] = '.' # over- or underpopulation
```

### Special Pythagorean Triplet (Problem 9)

A Pythagorean triplet is a set of three natural numbers $a < b < c$, for which $a^2 + b^2 = c^2$.
For example, $3^2 + 4^2 = 5^2$. There exists exactly one Pythagorean triplet for which $a + b + c = 1000$.
Find the product $abc$.

```python
x = 1000
for a in range(1,333):
    for b in range(a+1,((x-a)//2)):
        c = x - a - b
        if (a**2 + b**2) == c**2:
            break
print ([a,b,c])
```
