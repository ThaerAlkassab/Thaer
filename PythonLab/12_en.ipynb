## [matplotlib](https://matplotlib.org/)

- Matplotlib is a data visualization library.
- It is used by many other packages (e.g. pandas).
- It borrowed many ideas from MATLAB.
- It is strongest at static, 2D visualizations.

```python
# Import matplotlib and numpy!
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plotting the sin() function.
x = np.arange(-5,5,0.1)
y = np.sin(x)
plt.plot(x,y)
pass
```

```python
# Let's customize the previous plot!
# - increase the figure size
plt.figure(figsize=(12,4)) # creat new figure
# - increase the line width
plt.plot(x,y, lw=10)
# - display a grid
plt.grid(True)
# - add labels to the x and y axis
plt.xlabel('x')
plt.ylabel('sin x') # we can use LaTeX $\\......$ so we can add font formation
```

```python
# doing the same via pandas
import pandas as pd
pd.Series(y, index=x).plot(figsize=(12,4), lw=5, grid=True, xlabel='$x$', ylabel='$\\sin x$')
```

```python
plt.figure(figsize=(12,4)) # creat new figure
x = np.arange(-10,10,0.1)

plt.plot(x,np.sin(x), lw=5)
plt.plot(x,np.cos(x), lw=5)

plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend(['sin x', 'cos x'])
```

```python
plt.figure(figsize=(12,4)) # creat new figure
x = np.arange(-10,10,0.1)

plt.plot(x,np.sin(x), lw=5, label='sin x')
plt.plot(x,np.cos(x), lw=5, label='cos x')

plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend()
```

```python
# the same using DataFrame
df = pd.DataFrame({'$\\sin x$': np.sin(x), '$\\cos x$': np.cos(x)}, index=x)
df.plot(figsize=(12,4), lw=2, grid=True, xlabel='$x$')
```

```python
# Scatter plot.
n = 2000
x1 = np.random.normal(loc=0, size=n)
y1 = np.random.normal(loc=0, size=n)
x2 = np.random.normal(loc=2, size=n)
y2 = np.random.normal(loc=2, size=n)
plt.figure(figsize=(8, 8))
plt.scatter(x1, y1, alpha=0.33)
plt.scatter(x2, y2, alpha=0.33)
plt.grid(True)
```

```python
# Subplots.
x = np.arange(-10,10,0.1)
plt.plot(x,np.sin(2*x))
plt.plot(x,np.sin(3*x))
plt.plot(x,np.sin(4*x))
plt.plot(x,np.sin(5*x))
plt.plot(x,np.sin(6*x))
```

```python
plt.figure(figsize=(10,12))
for k in range(1,7):
    plt.subplot(3,2,k)
    plt.plot(x,np.sin(k*x))
    plt.title(f'$\\sin({k}x$)')
```

```python
# Saving a figure to file.
n = 2000
x1 = np.random.normal(loc=0, size=n)
y1 = np.random.normal(loc=0, size=n)
x2 = np.random.normal(loc=2, size=n)
y2 = np.random.normal(loc=2, size=n)
plt.figure(figsize=(8, 8))
plt.scatter(x1, y1, alpha=0.33)
plt.scatter(x2, y2, alpha=0.33)
plt.grid(True)
plt.savefig('Scater.png', bbox_inches='tight') # bbox to cut the unnecessary empty space around the figure
```

```python
# Annotations.
n = 2000
x1 = np.random.normal(loc=0, size=n)
y1 = np.random.normal(loc=0, size=n)
x2 = np.random.normal(loc=2, size=n)
y2 = np.random.normal(loc=2, size=n)
plt.figure(figsize=(8, 8))
plt.scatter(x1, y1, alpha=0.33)
plt.scatter(x2, y2, alpha=0.33)
plt.grid(True)

plt.annotate(
    'typical apple', xy=(0, 0), xytext=(1, -2.5), # add text with the arrow and set the start and the end of the arrow
    arrowprops=dict(arrowstyle='->', connectionstyle='angle3')
)
plt.annotate('typical banana', xy=(2, 2), ha='center') # add text inside
```

```python
# Pie chart plot.
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.0f%%', shadow=True)
pass
```
